```python
df = pd.read_sql_query(
    "select dc.CustomerKey, dc.YearlyIncome, dc.MaritalStatus, dc.Gender, dc.TotalChildren,\
        dc.NumberChildrenAtHome as ChildrenAtHome, dc.EnglishEducation as Education,\
        dc.EnglishOccupation as Occupation, dc.HouseOwnerFlag as HouseOwner,\
        dc.NumberCarsOwned as CarsOwned, 2012-YEAR(dc.BirthDate) as Age,\
        SUM(fis.SalesAmount) as TotalSpent\
    from DimCustomer dc\
    join FactInternetSales fis on fis.CustomerKey=dc.CustomerKey\
    group by dc.CustomerKey, dc.YearlyIncome, dc.MaritalStatus, dc.Gender, dc.TotalChildren,\
        dc.NumberChildrenAtHome, dc.EnglishEducation, dc.EnglishOccupation,\
        dc.HouseOwnerFlag, dc.NumberCarsOwned, 2012-YEAR(dc.BirthDate)",
    connection) 

df.head()
```

```python
df.index = df.pop('CustomerKey')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Boxplots:
def plot_boxplot(df, rows=5):
    fig, axs = plt.subplots(figsize=(20, 20))
    for index, col in enumerate(df.select_dtypes(include=[np.number]).columns):
      plt.subplot(len(df.columns)/rows + 1, rows, index+1)
      sns.boxplot(data=df, order=[col], color="skyblue", orient='v')
    plt.subplots_adjust(hspace=.5, wspace = .5)
    plt.show()
```

```python
plot_boxplot(df, rows=3)
```

```python
def drop_outliers(df):
  Q1 = df.quantile(0.25)
  Q3 = df.quantile(0.75)
  IQR = Q3 - Q1
  df_final = df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR))).any(axis=1)]
  return df_final
```

```python
df = drop_outliers(df)
print(df.shape)
plot_boxplot(df, rows=3)
```

```python
for col in df.select_dtypes(include=[object]).columns:
    print(df[col].value_counts())
    print('='*40)
```

```python
columns_to_be_encoded = ['MaritalStatus', 'Gender', 'Education', 'Occupation', 'HouseOwner']
```

# Mã hoá dữ liệu

```python
df_1 = pd.concat([df.drop(columns=columns_to_be_encoded),
                pd.get_dummies(df[columns_to_be_encoded])], axis=1)
df_1
```

# Chọn đặc trưng

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
target = 'TotalSpent'
```

```python
y = df_1[[target]]
X = df_1.drop(columns=[target])
```

```python
k = 8

best_features = SelectKBest(score_func=chi2, k=k)
fit = best_features.fit(X, y)

feature_scores = pd.concat([pd.DataFrame(X.columns), pd.DataFrame(fit.scores_)], axis=1)
feature_scores.columns = ['Feature', 'Score']
feature_scores.nlargest(k, 'Score')
```

```python
selected_features = feature_scores.nlargest(k, 'Score')['Feature'].values
selected_features
```

# Xây dựng mô hình

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
X_train, X_test = train_test_split(df_1[np.append(selected_features, target)],
                                   train_size=.7,
                                   test_size=.3,
                                   random_state=1)
y_train = X_train.pop(target)
y_test = X_test.pop(target)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Linear Regression

```python
lr = LinearRegression()
lr_model = lr.fit(X_train, y_train)
```

```python
lr_model.score(X_train, y_train)
```

```python
lr_y_pred = lr_model.predict(X_test)
lr_y_pred
```

```python
plt.figure(figsize=(15,8), dpi=80)
plt.plot(y_test.values)
plt.plot(lr_y_pred, color='red')
plt.show()
```

```python
mse = np.mean((lr_y_pred - y_test)**2)
'{:f}'.format(mse)
```

## Ordinary Least Squares

```python
X_train = sm.add_constant(X_train)
sm_lr = sm.OLS(y_train, X_train).fit()
print(sm_lr.params)
print(sm_lr.summary())
```

```python
X_test = sm.add_constant(X_test)
y_pred = sm_lr.predict(X_test)
y_pred
```

```python
plt.figure(figsize=(15,8), dpi=80)
plt.plot(y_test.values)
plt.plot(y_pred.values, color='red')
plt.show()
```

```python
mse = np.mean((y_pred - y_test)**2)
'{:f}'.format(mse)
```

## Đánh giá bằng K-fold

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
kfcv = KFold(n_splits=25, random_state=1, shuffle=True)
cv_model = LinearRegression()
lr_scores = cross_val_score(cv_model, X[selected_features], y, cv=kfcv, n_jobs=1)
print(f"Accuracy: mean={np.mean(lr_scores)}, std={np.std(lr_scores)}")
```
